# model-serving-stack — Demo
Quick walkthrough of vLLM OpenAI endpoint, Triton client, and Ray Serve router.

```python
from openai import OpenAI
client = OpenAI(base_url='http://localhost:8010/v1', api_key='unused')
response = client.chat.completions.create(
    model='llama3-8b',
    messages=[{'role': 'user', 'content': 'What is DCGM and why does it matter for GPU autoscaling?'}],
    max_tokens=256
)
print(response.choices[0].message.content)
```

## Triton Client Example

```python
from triton.client import TritonClient
tc = TritonClient(url='localhost:8000', protocol='http')
print('Triton ready:', tc.health_check())
result = tc.infer('llama3', prompt='Explain vLLM continuous batching.')
print(result)
```
